Criar a conexão com o BD, e a classe que representa os metadados, tabela no bd, inpeciona a estrutura do BD

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

Para criar o BD. Vamos usar um BD próprio do google colab

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

Passando o arquivo para um bd

```python
dados.to_sql('clientes', engine, index=False)
```

Para inspecionar, verificar o que tem no bd

```python
inspector = inspect(engine)
```

Para mostrar as tabelas que tem no bd

```python
print(inspector.get_table_names())
```

Como fazer uma consulta em sql

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

Para ler uma consulta sql

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

Para salvar essa consulta como uma nova tabela

```python
empregados.to_sql('empregados', con=engine, index=False)
```

Para ler uma tabela inteira

```python
pd.read_sql_table('empregados', engine)
```

Consultar determinadas colunas

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual'])
```

Seleciona todas as colunas da tabela cliente

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

Deletar uma informação
e o connect para estabelecer uma conexão com o bd e executar a query

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```

Para atualizar uma informação

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```
