```python
import pandas as pd
```

Os dados abaixo estão aninhados, em dicionários.

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

Desagrupando, podemos ver que da coluna descricao_local até a última as informações estão em listas

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

Para visualizar o nome de todas as colunas

```python
colunas = list(dados.columns)
colunas
```

Tirar as informações de listas

```python
dados = dados.explode(colunas[3:])
dados
```

Resetar os índices, que estão repetidos e criar uma nova coluna com índices numerados

```python
dados.reset_index(inplace = True, drop = True)
dados.head()
```

Para saber quais são os tipos de dados de cada coluna

```python
dados.info()
```

Para transformar os tipos de dados

```python
import numpy as np
```

Para converter os dados tipo object em inteiro

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

Tentando transformar a coluna preco para float, não dá certo porque tem um $, e não dá pra ter um caractere junto com o inteiro. Primeiro temos que tratar esses dados

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

Retorna uma cópia da coluna preco, sem o $ e sem vírgulas

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

Agora sim, dá para transformar em float

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

Retirando o $, e as vírgulas das duas colunas abaixo

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

Transformando todas as letras em minúsculo, da coluna descricao_local

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

Removendo caracteres que não precisam para a análise. Primeiro, o texto abaixo mostra com o scaracteres. Na célula seguinte, depois de retirados

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
```

Retirando os hífens que não fazem parte de uma palavra composta

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex = True)
```

split transforma os elementos textuais em lista

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

Removendo as {} e aspas que tem na coluna comodidades

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

Transformandod em minúsculas

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```
